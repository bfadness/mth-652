```python
import numpy as np
import matplotlib.pyplot as plt
```

Consider the simple ODE
$$
    u' = -u
$$
with initial condition $u(0) = 1$.
The exact solution is $u(t) = \exp(-t)$.

```python
t_f = 10
t = np.linspace(0, t_f, 100)
plt.plot(t, np.exp(-t))
```

We can use forward Euler and backward Euler to approximate the solution to this equation.

```python
u_fe = [1]
u_be = [1]
n = 100
dt = t_f/n

t_fe = np.linspace(0, t_f, n + 1)

for i in range(n):
    u_fe.append((1-dt)*u_fe[-1])
    u_be.append(u_be[-1]/(1+dt))
```

```python
plt.plot(t, np.exp(-t), "k", label="Exact")
plt.plot(t_fe, u_fe, ".", label="Forward Euler")
plt.plot(t_fe, u_be, ".", label="Backward Euler")
plt.legend()
```

However, if the condition
$$
    \Delta t \leq 2
$$
is not satisfied, the forward Euler solution will "blow up" instead of decaying exponentially.

```python
u_fe = [1]

dt = 2.1
t_fe = np.arange(0, t_f + dt, dt)

for i in range(len(t_fe) - 1):
    u_fe.append((1-dt)*u_fe[-1])

plt.plot(t, np.exp(-t), "k", label="Exact")
plt.plot(t_fe, u_fe, ".-", label="Forward Euler")
plt.legend()
```

The backward Euler method, on the other hand, remains stable for arbitrarily large $\Delta t$.

```python
u_be = [1]

dt = 2.1
t_be = np.arange(0, t_f + dt, dt)

for i in range(len(t_be) - 1):
    u_be.append(u_be[-1]/(1+dt))

plt.plot(t, np.exp(-t), "k", label="Exact")
plt.plot(t_fe, u_fe, ".-", label="Forward Euler")
plt.plot(t_be, u_be, ".-", label="Backward Euler")
plt.legend()
```
